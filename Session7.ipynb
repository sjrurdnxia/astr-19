```python
import math
import numpy as np
import matplotlib.pyplot as plt

x = []
i = 0
count = 2/100
while i <= 2:
    i += count
    x.append(round(i, 2))

```

```python
def exp():
    a = 5
    b = 2
    c = 1

    y = a * np.exp(-b * np.array(x) + c)

```

```python
exp()

plt.plot(x, y)
plt.xlabel(f'Time [milliseconds]')
plt.ylabel(f'Awesomeness')
plt.title('Exponential Curve')
plt.savefig("myImagePDF.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
